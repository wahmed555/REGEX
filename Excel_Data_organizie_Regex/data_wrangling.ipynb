```python
# seprate List  of  Store and Shops


StoreNBrand=[]
Address1=[]
Phone2=[]
for i in range(0,len(df1a),1):
    k=df1a.iloc[i,0]
    if (len(re.findall(patern_uppercase,k))==1):  # regex to check Name of Store Starts with Capital
        print(k)
        StoreNBrand.append(k)
        
        Address1.append(df1a.iloc[i+1,0])
        
        p=df1a.iloc[i+2,0]         # picking up location
        print(p)
        q=(re.findall(pattern_phone,p) )
        print(q)
        if (len(q)==1):
            Phone2.append(p)
        else:
            Phone2.append('-No Phone')
            
    
    else:
        print('Missed')
    
# Phons
```
